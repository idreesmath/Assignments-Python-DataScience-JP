Q4: Write a Python program which accepts the radius of a circle from the user and compute the area.

```python
radius = float(input("Enter Radius of Circle: "))
PI = 3.1415
area = PI * radius**2

print(f"The Area of Circle with radius {radius} is {area}")
```

Q5: Write a Python program which accepts the user's first name and print them in reverse order with space between them.

```python
first_name = input("Enter your First Name: ")
last_name = input("Enter your Last Name: ")

full_Name = first_name + " " + last_name

print(f"The Full Name is reverse order is {full_Name[::-1]}")
```
